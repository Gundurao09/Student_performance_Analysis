```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# 1. Bar Chart: Average CGPA by Department
avg_cgpa_dept = data1.groupby('Department')['CGPA'].mean()
axs[0, 0].bar(avg_cgpa_dept.index, avg_cgpa_dept.values, color='skyblue')
axs[0, 0].set_title("Average CGPA by Department")
axs[0, 0].set_ylabel("Average CGPA")
axs[0, 0].set_xlabel("Department")

# 2. Scatter Plot: Attendance vs CGPA
colors = {'CSE': 'red', 'ECE': 'blue', 'EEE': 'green', 'ME': 'orange', 'CIVIL': 'purple'}
color_list = data1['Department'].map(colors)
axs[0, 1].scatter(data1['Attendance (%)'], data1['CGPA'], c=color_list)
axs[0, 1].set_title("Attendance vs CGPA")
axs[0, 1].set_xlabel("Attendance (%)")
axs[0, 1].set_ylabel("CGPA")

# 3. Bar Chart: Count by Placement Status
placement_counts = data1['PlacementStatus'].value_counts()
axs[1, 0].bar(placement_counts.index, placement_counts.values, color='lightcoral')
axs[1, 0].set_title("Count by Placement Status")
axs[1, 0].set_ylabel("Count")

# 4. Box Plot: CGPA by Sports Participation
avg_sports = df.groupby('Sports Participation')['CGPA'].mean()
axs[1, 1].bar(avg_sports.index, avg_sports.values, color='lightgreen')
axs[1, 1].set_title("Average CGPA by Sports Participation")
axs[1, 1].set_ylabel("Average CGPA")
axs[1, 1].set_xlabel("Sports Participation")

# Layout adjustment
plt.tight_layout()
plt.show()
```
